```python
import matplotlib.pyplot as plt
```

```python
month = ['Jan','Feb','Mar','Apr','May','Jun']
salary = ['60,000','60,000','60,000','70,000','70,000','85,000']
```

```python
#set font
font = {'family':'Georgia','color':'B','size':20}
```

```python
#title
plt.title("simple line chart",fontdict = font)
#Axis label
plt.xlabel("Month",fontdict=font)
plt.ylabel("Salary",fontdict=font)
plt.plot(month,salary,color='M',linestyle='solid',linewidth=10,markerfacecolor="tab:blue",markersize=10)
plt.show()
#plt.savefig('line.jpeg')
```
