```python
title = imdb_movie.groupby('Series_Title').sum()
title
```

```python
movie_genre = imdb_movie.groupby('Genre').sum()
movie_genre
```

```python
#Print the highest rated movies list
highest_rated_movie = title.query("IMDB_Rating > 8.5 & IMDB_Rating < 9.0")
highest_rated_movie
```

```python
title_genre = imdb_movie[['Series_Title','Genre','IMDB_Rating']]
title_genre
```

```python
#To check particular string in column
imdb_movie[imdb_movie['Genre'].str.contains('Action')]
```

```python
#splitting genre column
title_genre['Drama_list'] = title_genre['Genre'].apply(lambda x : x.split(',')[0])
title_genre.head()
```
